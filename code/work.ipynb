```python
df_map = file[["LONGITUDE","LATITUDE"]]
BBox = ((df_map.LONGITUDE.min(), df_map.LONGITUDE.max(), df_map.LATITUDE.min(), df_map.LATITUDE.max()))
ruh_m = plt.imread('map.png')
fig, ax = plt.subplots(figsize = (8,7))
ax.scatter(df_map.LONGITUDE, df_map.LATITUDE, zorder=1, alpha= 0.2, c='b', s=10)
ax.set_title('Plotting')
ax.set_xlim(BBox[0],BBox[1])
ax.set_ylim(BBox[2],BBox[3])
ax.imshow(ruh_m, zorder=0, extent = BBox, aspect= 'equal')
```

```python
df=file[["STATION","DATE","TEMP"]]


upper = df['TEMP'].mean() + 3*df['TEMP'].std()
lower=df['TEMP'].mean() - 3*df['TEMP'].std()

#newdf = df[(df['MAX']<max_upper) & (df['MAX']>max_lower)&(df['MIN']<min_upper) & (df['MIN']>min_lower)]
#newdf=newdf.assign(DIFF = abs(newdf['MAX'] - newdf['MIN']))
newdf = df[(df['TEMP']<upper) & (df['TEMP']>lower)]
newdf=newdf.groupby('STATION').mean()
newdf[newdf["TEMP"]==newdf["TEMP"].max()]
```

```python
spec = file[file['STATION']==85987199999]
spec=spec.assign(YEAR = spec['DATE'].str[:4])
spec.drop('DATE', axis=1, inplace=True)
spec=spec.groupby('YEAR').mean()
spec.plot(y=['TEMP','MIN','MAX'], kind='line')
```
