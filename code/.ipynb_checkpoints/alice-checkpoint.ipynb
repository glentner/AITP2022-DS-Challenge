```python
def plot_stations(df, *stations):
    fig = px.scatter_geo(df, lat= 'LATITUDE', lon="LONGITUDE", scope = 'world', width = 800, height = 600)
    fig.update_layout(geo = dict(projection_scale=7, center=dict(lat=df['LATITUDE'].iloc[0], lon=df['LONGITUDE'].iloc[0])))
    fig.plot()
```

```python
plot_stations(mydat, 1)
```
