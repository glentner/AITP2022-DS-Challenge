```python
df = data[['STATION', 'MAX', 'MIN', 'DATE']]
df = df.assign(DIFF = abs(df['MAX'] - df['MIN']))
df[df['DIFF'] == df['DIFF'].max()]
```

```python
q6 = data[data['STATION'] == 89016099999]
q6 = q6.assign(YEAR = data['DATE'].str[:4])
q6 = q6.groupby('YEAR').mean()
q6
```

```python
twotwo = data[data['DATE'] >= '2022-01-10']
twotwo = twotwo.assign(YEAR = data['DATE'].str[:4])
twotwo.plot('DATE', 'TEMP', rot=45)
```
